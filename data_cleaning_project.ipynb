```python
#Importing libraries
import numpy as np
import pandas as pd
```

```python
# reading data
df = pd.read_csv('/workspaces/data_cleaning_project/data/Cafe Sales.csv')
```

```python
df.head()
```

```python
#inspecting last few records
df.tail()
```

```python
#data structure
df.info()
```

```python
#duplicate records
print(f'Number of duplicated records : {sum(df.duplicated())}')
```

```python
# missing records after replacing Error and Unknown
df.isna().sum() * 100/df.shape[0]
```

```python
# unique records
df_columns = set(df.columns)
exclude = {'Transaction ID'}
df_columns = df_columns.difference(exclude)

for column in df_columns:
    print(f'{column}:\n{df[column].unique()}',end = '\n\n')
```

```python
#replacing ERROR and Unknown values
df = df.replace(['ERROR','UNKNOWN'],np.nan)
```

```python
df.info()
```

```python
df['Quantity'] = df['Quantity'].astype('Int64')
df[['Price Per Unit','Total Spent']] = df[['Price Per Unit','Total Spent']].astype('float')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
df.info()
```

```python
#dropping records where Transaction date is not available
df = df.dropna(subset = ['Transaction Date'])
```

```python
df.info()
```

```python
# replacing missing records in payment 
df[['Payment Method','Location']] = df[['Payment Method','Location']].fillna('INA')
```

```python
df['Item']= df['Item'].fillna(df['Item'].mode()[0])
```

```python
df.describe()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
df['Quantity'].median()
```

```python
df.groupby('Item')['Price Per Unit'].value_counts()
```

```python
df['Price Per Unit'].mode()
```

```python
#filling missing records in price per unit with mode value of item category
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(lambda x: x.fillna(x.mode()[0]) )
```

```python
df.info()
```

```python
#filling missing records in Total Spent
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'])*df['Price Per Unit']
```

```python
df.info()
```

```python
# saving clean data
df.to_csv('/workspaces/data_cleaning_project/data/Cafe Sales (Cleaned).csv',index=False)
```
